## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
#rawdata
```

```python
df1 = rawdata[0]
df1
```

```python
df2 = rawdata[1]
df2
```

```python
#merge two dfs of physical best sellers
df_physical = pd.concat([df1,df2])
df_physical
```

```python
# Add a new column to indicate physical copies
df_physical['type'] = "physical"
df_physical
```

```python
df3 = rawdata[3]
df3
```

```python
df4 = rawdata[4]
df4
```

```python
#merge two dfs of digital best sellers
df_digital = pd.concat([df3,df4])
df_digital
```

```python
# Add a new column to indicate digital copies
df_digital['type'] = "digital"
df_digital
```

```python
#merge physical and digital dfs together
df_all = pd.concat([df_physical,df_digital], ignore_index=True) #ignore index so that it is continuous for all
df_all
```

```python
#best singles of 2003 cause there are none in 2004
bestof2003 = df_all[df_all["Released"] == 2003]
bestof2003
```

```python
#best singles of 2018, when I was 15 years old
bestof2018 = df_all[df_all["Released"] == 2018]
bestof2018
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
#rawdata
```

```python
df_space1950 = rawdata[0]
df_space1950
```

```python
df_space2020 = rawdata[7]
df_space2020
```

```python
#merge 1950 and 2020 dfs together
df_spaceall = pd.concat([df_space1950,df_space2020], ignore_index=True)
df_spaceall
```

```python
#output most common date (take only last four characters from Launch date to only use year)
df_spaceall['Launch date'].str[-4:].value_counts().nlargest(1)
```

```python
#output 5 most common dates (take only last four characters from Launch date to only use year)
df_spaceall['Launch date'].str[-4:].value_counts().nlargest(5)
```

```python
#before getting the most common month for space launches
# Convert the string to datetime
df_spaceall['Launch date'] = pd.to_datetime(df_spaceall['Launch date'])

# Extract the month from the datetime column
df_spaceall['Month'] = df_spaceall['Launch date'].dt.month
print(df_spaceall['Month'])
```

```python
#get the most common month for space launches
df_spaceall['Month'].value_counts().nlargest(1)
```

```python
#same can be done for year this way
# Extract the year from the datetime column
df_spaceall['Year'] = df_spaceall['Launch date'].dt.year
print(df_spaceall['Year'])
```

```python
#get the most common month for space launches
df_spaceall['Year'].value_counts().nlargest(1)
```

```python
#get the all months in order most to least launches (remove: nlargest(1))
df_spaceall['Month'].value_counts()
```

## Supervillain challenge

1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
```

```python
#df_villain1 = rawdata[3] 
#df_villain2 = rawdata[4] 
#instead of one by one, loop through all

# Create an empty list to store the individual dataframes
df_villains = []

# Loop through the rawdata list starting from index 3 until the end
for x in range(3, len(rawdata)):  # len(rawdata) - the total number of elements
    df_villain = rawdata[x]       # Get each dataframe from the list
    df_villains.append(df_villain)        # Append it to the list

# Combine all dataframes
df_villains = pd.concat(df_villains, ignore_index=True)
df_villains
```

```python
# Create new column just with the first four characters (year)
df_villains['Year'] = df_villains['Year Debuted'].str.slice(0, 4)

# Print the new year column
print(df_villains['Year'])
```

```python
# clean up the year column a bit
# The pattern \D looks for any character that is not a digit
non_numeric_years = df_villains[df_villains['Year'].str.contains(r'\D', na=False)]

# Display the rows where the 'Year' contains non-numeric characters
print(non_numeric_years)

# look for those that have more than a date and replace with the right date
```

```python
# replace with the right date
# Replace wrong values based on conditions
df_villains['Year'] = df_villains['Year'].replace('(Faw' ,1943)
df_villains['Year'] = df_villains['Year'].replace('Mong' ,1980)
df_villains['Year'] = df_villains['Year'].replace('Curt' ,1987)
df_villains['Year'] = df_villains['Year'].replace('Blaz' ,1990)

#this doesn't work for this one for some reason
#df_villains['Year'] = df_villains['Year'].replace("#1:" ,1976)
```

```python
# Verify location of "#1:"
print(df_villains.iloc[373, 5])
```

```python
# Replace the value at row 373 and column 5 with 1976
df_villains.iloc[373, 5] = '1976'
```

```python
# Create new column just with Decade
df_villains['Decade'] = df_villains['Year'].str.slice(0, 3) + "0s"

# Print the decade column
print(df_villains['Decade'])
```

```python
#get the all decades in order most to least villains
df_villains['Decade'].value_counts()
```

```python
#get the all companies in order most to least villains
df_company = df_villains['Company'].value_counts()
df_company
```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df.head()
```

```python
# Transforming the data from wide format to long format
df_long = df.melt(id_vars=["Year"], 
                  var_name="Task", 
                  value_name="RT")
df_long
```

```python
# adding participant id (same as index)
df_long = df_long.reset_index().rename(columns={"index": "Participant_id"})
df_long
```

## Stroop challenge 2 (Advanced!!!):

Make a new dataframe which shows the mean response time (in seconds) for each task for each year.

```python
# Creating a new dataframe showing the mean response time for each task for each year
df_mean_rt = df_long.groupby(['Year', 'Task']).agg(mean_RT=('RT', 'mean')).reset_index()
df_mean_rt
```
